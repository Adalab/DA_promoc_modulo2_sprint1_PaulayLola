1.Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('attacks_pandas_5.csv')
```

```python
df.head()
```

```python
df_2015 = df[df['year'] >= 2015.0]
df_2015.head(2)
```

```python
plt.figure(figsize =(12,10))

fig = sns.countplot(y='activity',data = df_2015, hue ='year')

fig.set(xlabel='Número de personas', ylabel = 'Actividad')

fig.set_title('Personas por actividad desde 2015')

plt.legend(bbox_to_anchor=(1.05, 1))

plt.show()
```

Deducimos que sufren más ataques las personas que hacen surf o nadando debido a posibles patadas a los animales y la llamada de atención provocada por el braceo y las salpicaduras. Que este resultado se dispare en 2015 puede ser debido al aumento de la temperatura del agua, haciendo que se los tibuones muevan más para buscar mejor entorno. La caída del número de ataques en años sucesivos puede deberse a las medidas de segurida tomadas para evitar edtos accidentes. Existen dipositivos disuasorios o neoprenos con estampados para no ser confundios con su presas.

3.¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.

```python
df_1500 = df[df['year'] >= 1500.0]
df_1500.head(2)
```

```python
plt.figure(figsize =(27,4))

fig = sns.countplot(x='year',data = df_1500)

fig.set(xlabel='Año', ylabel = 'Ataques')
plt.xticks(rotation = 90)
fig.set_title('Ataques por año desde 1500')
```

En la década de los 60 se distingue un ligero aumento en el número de ataques debido al boom del surf como deporte en auge. Estas cifras decaen gradualmente al empezar a existir mayor precacución y medidas de seguridad. A partir de los años 80 se mantienen los datos en 20 casos aproximadamente y es a partir del siglo XXI que se ve un considerble aumento de casos debido al aumento de la tempertura del agua y la proliferación de múltiplpes deportes acuáticos.  En 2015 observamos un pico que comienza a descender porque las medidas de seguridad están más exrendidas y el miedo o precacución de los bañistas. Existe mayor concienciación ante este peligro.

4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
usa_aus_sa =df[df['country'].isin(['USA','AUSTRALIA', 'SOUTH AFRICA'])]
usa_aus_sa.head(2)
```

```python
usa_aus_sa_1900 = usa_aus_sa[usa_aus_sa['year'] >=1900.0]# filtramos por año 
usa_aus_sa_1900.head(2)
```

```python
plt.figure(figsize =(14,9))

fig = sns.countplot(x='country',hue = 'year', data = usa_aus_sa_1900)

fig.set(xlabel='Pais', ylabel = 'Ataques')

fig.set_title('Ataques por año desde 1900 en USA, Australia y Sudáfrica')
fig.legend(bbox_to_anchor =(1,1),ncol = 4)
```

En estos tres países debido a las condiciones marinas hay más presencia de tiburones que en otros países. En USA hay mayor número de población lo que indica que pudiera haber más ataques; destaca su popularidad en la zona de California, frecuentada también por muchos turistas. Los casos aumentan en este siglo por lo que comentamos en el apartado anterior. En el caso de Australia, el surf está muy presente en el día a día de sus habitantes. El caso curioso de Sudáfrica es que no se provoca el pico del siglo XXI por la disminución repentina debido la caza furtiva de tiburones y los ataques recibidos por parte de las orcas.

5. Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
df_boxplot = usa_aus_sa.groupby(['country','year'])['case_number'].count().reset_index()
df_boxplot
```

```python
plt.figure(figsize =(20,9))  # agrupamos los ataques por año de los países USA, Australia y Sudáfrica
fig =sns.boxplot(x = 'country', y = 'case_number', data = df_boxplot)

fig.set(xlabel='Pais', ylabel = 'Número de ataques')


```

Podemos deducir de este boxplot que en EEUU hay mayor número de ataques y hay tres años en el que esatas cifras se disparan.
En Sudáfrica las cifras son basrante uniformes, es decir no hay subida de número de ataques. Sin embargo en Australia, aunque hay un menor numero de ataques en general, se observa una fuerte subida en cinco años concretamente. Los años no se muestran en esta gráfica, pero se aprecian en la anterior.
