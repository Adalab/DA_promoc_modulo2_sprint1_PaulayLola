4.Filtrad los datos para quedarnos solo con los datos de USA.
- ¿Cuántos ataques de tiburones hubo en USA a lo largo del tiempo? Muestra solo los que hayan ocurrido del 2000 en adelante.
- En función del tipo, ¿cuántos ataques de tiburones de cada tipo hubo en USA? Muestra los resultados en un gráfico de línea.

```python
usa = df[df['country']=='USA']
usa.head()
```

```python
ataques_usa = pd.DataFrame(usa.groupby('year')['case_number'].count())
ataques_usa.reset_index(inplace=True)
ataques_usa.head()
```

```python
ataques_usa_2000 = ataques_usa[ataques_usa['year']>=2000.0]
ataques_usa_2000.head()
```

```python
plt.figure()

plt.bar(ataques_usa_2000["year"], ataques_usa_2000["case_number"], facecolor  = 'turquoise', edgecolor = "black" )

plt.show() 
```

En función del tipo, ¿cuántos ataques de tiburones de cada tipo hubo en USA? Muestra los resultados en un gráfico de línea.

```python
ataques_usa_tipo = pd.DataFrame(usa.groupby('type')['case_number'].count())
ataques_usa_tipo.reset_index(inplace=True)
ataques_usa_tipo.head()
```

```python
plt.figure()
plt.plot(ataques_usa_tipo['type'],ataques_usa_tipo['case_number'])
plt.show() 
```
